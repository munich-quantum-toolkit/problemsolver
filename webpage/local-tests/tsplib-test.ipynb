```python
import os
import os.path
import mqt.qubomaker as qm
from mqt.qubomaker import Graph
from mqt.qubomaker import pathfinder as pf
import tsplib95
```

```python
graph = Graph.from_adjacency_matrix([
    [0, 2, 6, 6, 2],
    [5, 0, 1, 7, 8],
    [7, 3, 0, 5, 4],
    [4, 8, 1, 0, 3],
    [9, 6, 7, 2, 0],
])
```

```python
encoding_type = pf.EncodingType.ONE_HOT
n_paths = 1
max_path_length = graph.n_vertices
loop = True
settings = pf.PathFindingQUBOGeneratorSettings(encoding_type, n_paths, max_path_length, loop)
```

```python
generator = pf.PathFindingQUBOGenerator(
    objective_function=pf.MinimizePathLength(path_ids=[1]), 
    graph=graph, 
    settings=settings
)
```

```python
generator.add_constraint(pf.PathIsValid(path_ids=[1]))
generator.add_constraint(pf.PathContainsVerticesExactlyOnce(vertex_ids=graph.all_vertices, path_ids=[1]))
generator.add_constraint(pf.PrecedenceConstraint(1, 3, path_ids=[1]))
generator.add_constraint(pf.PrecedenceConstraint(2, 1, path_ids=[1]))
generator.add_constraint(pf.PrecedenceConstraint(5, 4, path_ids=[1]))
```

```python
#generator.construct_expansion()
generator.graph.n_vertices
```

```python
matrix = generator.construct_qubo_matrix()
qm.print_matrix(matrix)
```

```python
print(generator.count_required_variables())
```

```python
encoding_type = pf.EncodingType.ONE_HOT
n_paths = 2
max_path_length = 3
loop = False
settings2 = pf.PathFindingQUBOGeneratorSettings(encoding_type, n_paths, max_path_length, loop)

generator2 = pf.PathFindingQUBOGenerator(
    objective_function=pf.MinimizePathLength(path_ids=[1, 2]), 
    graph=graph, 
    settings=settings2
)
```

```python
generator2.add_constraint(pf.PathIsValid(path_ids=[1, 2]))
generator2.add_constraint(pf.PathsShareNoVertices(1, 2))
generator2.add_constraint(pf.PathPositionIs(vertex_ids=[1], position=1, path=1))
generator2.add_constraint(pf.PathPositionIs(vertex_ids=[2], position=1, path=2))
generator2.add_constraint(pf.PathEndsAt(vertex_ids=[4], path=1))
generator2.add_constraint(pf.PathEndsAt(vertex_ids=[5], path=2))
```

```python
matrix = generator2.construct_qubo_matrix()
qm.print_matrix(matrix)
print(matrix.shape)
```
